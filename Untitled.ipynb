```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split

# Load the dataset
movies = pd.read_csv('dataset.csv')

# Select relevant columns
movies = movies[['id', 'title', 'overview', 'genre']]

# Combine 'overview' and 'genre' into 'tags'
movies['tags'] = movies['overview'] + movies['genre']

# Drop unnecessary columns
new_data = movies.drop(columns=['overview', 'genre'])

# Split the dataset into training and validation sets
train_data, validation_data = train_test_split(new_data, test_size=0.2, random_state=42)

# Vectorize the 'tags' column
cv = CountVectorizer(max_features=10000, stop_words='english')
vector = cv.fit_transform(train_data['tags'].values.astype('U')).toarray()

# Calculate cosine similarity
similarity = cosine_similarity(vector)

# Function to recommend movies for a user
def recommend_movies_for_user(movie_attributes):
    movie_row = new_data[
        (new_data['title'] == movie_attributes['title']) &
        (new_data['overview'] == movie_attributes['overview']) &
        (new_data['genre'] == movie_attributes['genre'])
    ]
    
    if not movie_row.empty:
        index = movie_row.index[0]
        distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])
        recommendations = [new_data.iloc[i[0]].title for i in distance[1:6]]
        return recommendations
    else:
        print(f"Movie not found.")
        return []

# Example test case: Liked movie attributes
liked_movie_attributes = {
    'id': 1,
    'title': 'The Dark Knight',
    'overview': 'A great movie overview.',
    'genre': 'Action, Crime, Drama'
}

# Generate recommendations for the liked movie
recommendations = recommend_movies_for_user(liked_movie_attributes)

# Evaluate the recommendations
correct_predictions = set([liked_movie_attributes['title']])
accuracy = len(set(recommendations).intersection(correct_predictions)) / len(correct_predictions)

print(f'Liked Movie: {liked_movie_attributes["title"]}')
print(f'Recommended Movies: {recommendations}')
print(f'Accuracy: {accuracy}')
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
movies = pd.read_csv('dataset.csv')

# Select relevant columns
movies = movies[['id', 'title', 'overview', 'genre']]

# Combine 'overview' and 'genre' into 'tags'
movies['tags'] = movies['overview'] + movies['genre']

# Drop unnecessary columns
new_data = movies.drop(columns=['overview', 'genre'])

# Vectorize the 'tags' column
cv = CountVectorizer(max_features=10000, stop_words='english')
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()

# Calculate cosine similarity
similarity = cosine_similarity(vector)

# Function to calculate precision at k
def precision_at_k(selected_id, k):
    index = new_data[new_data['id'] == selected_id].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])
    
    # Get the relevant items for the selected movie
    relevant_items = set(new_data[new_data['id'] == selected_id]['title'])
    
    # Get the recommended items
    recommended_items = [new_data.iloc[i[0]].title for i in distance[1:k+1]]
    
    # Calculate precision at k
    precision = len(set(recommended_items).intersection(relevant_items)) / k
    return precision

# Example: Calculate precision at k for a selected movie with id=123 and k=5
selected_id = 238
k = 5
precision = precision_at_k(selected_id, k)

print(f'Precision at {k} for Movie {selected_id}: {precision}')
```
